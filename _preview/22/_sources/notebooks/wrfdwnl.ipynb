```python
import requests
from bs4 import BeautifulSoup
import os

# Function to download files
def download_file(url, local_dir):
    filename = os.path.join(local_dir, url.split('/')[-1])
    with open(filename, 'wb') as f:
        response = requests.get(url)
        f.write(response.content)
    print(f"Downloaded {filename}")

# URL of the website to scrape
url = "https://iop.archive.arm.gov/arm-iop-file/2021/guc/sail/xu-wrf/wrf_east_river_simulation_2023_subset/SAIL_WRF_2023/6hrly/"

# Local directory to save files
local_dir = "./downloaded_files"

# Create directory if it doesn't exist
if not os.path.exists(local_dir):
    os.makedirs(local_dir)

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.text, "html.parser")

# Find all links in the page
links = soup.find_all("a")

# Loop through all links and download files
for link in links:
    href = link.get('href')
    if href.endswith('00'):  # Change the extension according to the file type you want to download
        file_url = url + href
        download_file(file_url, local_dir)
```
