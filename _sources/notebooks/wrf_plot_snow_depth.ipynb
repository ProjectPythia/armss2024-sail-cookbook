# Plot snow depth from WRF Simulations
WRF simulations serve as comparison to observations. Here the differences in snow depth are compared for the high dust and black carbon loading events. 

```python
import xarray as xr
import cartopy
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
%matplotlib inline
```

Below, the first two plots show the snow depth for each day under consideration. The third plot shows differences in snow depth for the two days under consideration. 

```python
# Plot time differences in snowdepth for black carbon event on January 25th 
# compare snowdepth on January 25th and January 30th
dataset_25 = xr.open_mfdataset('wrf_data/jan_2022/wrfhourly_d03_2022-01-25.nc')
dataset_30 = xr.open_mfdataset('wrf_data/jan_2022/wrfhourly_d03_2022-01-30.nc')
mean_dataset_diff = dataset_30.mean(dim="Time") - dataset_25.mean(dim="Time")

fig = plt.figure(figsize = (12,3),constrained_layout=True)

gs = GridSpec(1, 3, figure=fig)
ax1 = fig.add_subplot(gs[0, 0])
dataset_25.mean(dim="Time").SNOWH.plot(x='south_north',y='west_east',cbar_kwargs={'label': r"Snow Depth / m"})
ax1.set_title('Jan 25, 2022')
ax2 = fig.add_subplot(gs[0, 1])
dataset_30.mean(dim="Time").SNOWH.plot(x='south_north',y='west_east',cbar_kwargs={'label': r"Snow Depth / m"})
ax2.set_title('Jan 30, 2022')
ax3 = fig.add_subplot(gs[0, 2])
mean_dataset_diff.SNOWH.plot(x='south_north',y='west_east',cbar_kwargs={'label': r"Snow Depth / m"})
ax3.set_title('Time difference')
```

```python
# Plot time differences in snowdepth for black carbon event on January 25th, 2022
# compare snowdepth on January 25th and January 26th
dataset_25 = xr.open_mfdataset('wrf_data/jan_2022/wrfhourly_d03_2022-01-25.nc')
dataset_26 = xr.open_mfdataset('wrf_data/jan_2022/wrfhourly_d03_2022-01-26.nc')
mean_dataset_diff = dataset_26.mean(dim="Time") - dataset_25.mean(dim="Time")

fig = plt.figure(figsize = (12,3),constrained_layout=True)

gs = GridSpec(1, 3, figure=fig)
ax1 = fig.add_subplot(gs[0, 0])
dataset_25.mean(dim="Time").SNOWH.plot(x='south_north',y='west_east',cbar_kwargs={'label': r"Snow Depth / m"})
ax1.set_title('Jan 25, 2022')
ax2 = fig.add_subplot(gs[0, 1])
dataset_26.mean(dim="Time").SNOWH.plot(x='south_north',y='west_east',cbar_kwargs={'label': r"Snow Depth / m"})
ax2.set_title('Jan 26, 2022')
ax3 = fig.add_subplot(gs[0, 2])
mean_dataset_diff.SNOWH.plot(x='south_north',y='west_east',cbar_kwargs={'label': r"Snow Depth / m"})
ax3.set_title('Time difference')
```
